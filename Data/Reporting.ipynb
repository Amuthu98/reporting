```python
import pandas as pd
import numpy as np
import nltk
import os
import nltk.corpus
```

## Dataframe des 100 vid√©os ayant fait le plus de vues par cat√©gories

```python
categ_id = 45
df_top=pd.DataFrame()
for i in range (categ_id):
    df=df_fr.loc[df_fr['category_id']==i]
    df=df.sort_values(by = 'views',ascending=False)
    df_top100=df[:100]
    df_top=pd.concat([df_top, df_top100])
df_top
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist
```

```python
def recup_words(text):
    fr = set(stopwords.words('french'))
    token = word_tokenize(text.lower())
    goodwords= [x for x in token if x not in fr]
    eng = set(stopwords.words('english'))
    goodwords2= [x for x in goodwords if x not in eng]
    return goodwords2
```

## Fr√©quence des mots du dataframe des top vid√©os / cat√©grories

```python
liste_words=df_top['title'].apply(lambda x: recup_words(x))
liste_words=liste_words.tolist()
```

```python
allWords = []
for words in liste_words:
    for w in words:
        if len(w)>1:
            allWords.append(w)
freq=FreqDist(allWords)
```

```python
freq_top = freq.most_common (40) 
df_freq=pd.DataFrame(freq_top)
df_freq
```
